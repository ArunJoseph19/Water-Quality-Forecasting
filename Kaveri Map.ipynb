```python
import requests
import json
import polyline

def get_shortest_route(s,e):
    #print(s,e)
    #route_url='http://router.project-osrm.org/route/v1/driving/79.81039,12.00679;80.28150,13.08195?alternatives=true&geometries=polyline'
    #print('http://router.project-osrm.org/route/v1/driving/'+str(s[0])+','+str(s[1])+';'+str(e[0])+','+str(e[1])+'?alternatives=true&geometries=polyline')
    route_url='http://router.project-osrm.org/route/v1/driving/'+str(s[0])+','+str(s[1])+';'+str(e[0])+','+str(e[1])+'?alternatives=true&geometries=polyline'
    r=requests.get(route_url)
    res=r.json()
    #print(res['routes'][0]['geometry'])
    routes=polyline.decode(res['routes'][0]['geometry'])
    print(float(res['routes'][0]['distance']))
    return routes,float(res['routes'][0]['distance'])
```
